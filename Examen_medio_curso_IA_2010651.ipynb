```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Leer el archivo CSV desde la URL
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv"
df = pd.read_csv(url)
```

```python
# Mostrar las primeras 5 filas
print(df.head())
```

```python
# Mostrar los tipos de datos de cada columna
print(df.dtypes)
```

```python
# Verificar valores nulos en la columna horsepower
print(df['horsepower'].isnull().sum())
```

```python
# Reemplazar valores nulos con la mediana
mediana_hp = np.median(df['horsepower'].dropna())
df['horsepower'] = df['horsepower'].fillna(mediana_hp)
```

```python
# Crear nueva columna relacion_peso_potencia
df['relacion_peso_potencia'] = df['weight'] / df['horsepower']
```

```python
# Histograma: distribución de aceleración
plt.figure()
plt.hist(df['acceleration'], bins=15)
plt.title("Distribución de la aceleración de los carros")
plt.xlabel("aceleracion")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Gráfico de dispersión: displacement vs horsepower
plt.figure()
plt.scatter(df['displacement'], df['horsepower'])
plt.title("Desplazamiento del motor vs Caballos de fuerza")
plt.xlabel("Displacement")
plt.ylabel("Horsepower")
plt.show()
```

```python
# Gráfico de barras: promedio de mpg por país de origen
mpg_promedio = df.groupby('origin')['mpg'].mean()
plt.figure()
mpg_promedio.plot(kind='bar')
plt.title("Promedio de MPG por pais")
plt.xlabel("Pais de origen")
plt.ylabel("Promedio MPG")
plt.show()
```
